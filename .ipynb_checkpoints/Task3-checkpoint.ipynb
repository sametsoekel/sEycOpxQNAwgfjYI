```r
library(fst)
library(data.table)
library(tidyverse)
library(lubridate)
library(modeltime)
library(tidymodels)
library(rsample)
library(modeltime.h2o)
library(gridExtra)
options(repr.plot.width=15)
options(scipen=999)
```

# Aggregating Data

I will use the data which I created at Task1, it includes all variables. 

Just going to joing `live_campaigns`

```r
df <- read.fst('data_serialized/aggregate_data.fst',as.data.table=T)
```

```r
campaigns <- read.fst('data_serialized/live_digital_campaigns.fst',as.data.table=T)
```

```r
df <- left_join(df,mutate(campaigns,date=as.Date(date)),by='date')
```

## Aggregating Daily

```r
df2 <- df[,.(visitor_num = sum(visitor_num),usdtry = mean(usdtry),
      eurtry = mean(eurtry), eurusd = mean(eurusd), faiz = mean(faiz), live_campaign = mean(live_campaign)),
  by = c('date','ga_channels')]
```

## Creating Variables

```r
df2[,wday := wday(date)]
df2[,yday := yday(date)]
df2[,saturday := +(wday==6)]
df2[,sunday := +(wday==7)]

df2[is.na(ga_channels),ga_channels:='Other']
```

## Splitting Data For Each Channel

```r
splitted_data <- df2 %>%
group_by(ga_channels) %>% 
group_split()

crm <- splitted_data[[1]]
direct <- splitted_data[[2]]
facebook <- splitted_data[[3]]
googleads <- splitted_data[[4]]
search <- splitted_data[[5]]
other <- splitted_data[[6]]
referral <- splitted_data[[7]]

rm(splitted_data)
gc()

setDT(crm)
setDT(direct)
setDT(facebook)
setDT(googleads)
setDT(search)
setDT(other)
setDT(referral)
```

## Adding shift for each channel

```r
crm[,shift := lag(visitor_num)]
direct[,shift := lag(visitor_num)]
facebook[,shift := lag(visitor_num)]
googleads[,shift := lag(visitor_num)]
search[,shift := lag(visitor_num)]
other[,shift := lag(visitor_num)]
referral[,shift := lag(visitor_num)]
```

## Note : Since I want to use all variables and the others (economic variables etc.) are not as old as transactions dataset, I am working a littler sample.

# Modeling

```r
tablenames <- c('crm','direct','facebook','googleads','search','other','referral')
```

## Fitting Gradient Boosting Models (boosts ARIMA's errors by XGBoost)

```r
set.seed(1923)

for(i in tablenames){
    eval(parse(text=sprintf('%s_split <- initial_time_split(%s,0.7)',i,i)))
    eval(parse(text=sprintf('%s_model <- fit(set_engine(automl_reg(mode="regression"),engine="h2o",max_runtime_secs = 20,nfolds=0,seed=2025),visitor_num ~ live_campaign + faiz + usdtry + eurtry + eurusd + shift + date + wday + saturday + sunday + yday,training(%s_split))',i,i)))
}
```

```r
for(i in tablenames){
    eval(parse(text = sprintf('%s_plot_test <- ggplot(cbind(testing(%s_split),predict(%s_model,testing(%s_split))),aes(date,visitor_num)) + geom_line(color="red")+ geom_line(aes(date,.pred),color="blue") + ggtitle("%s")',i,i,i,i,toupper(i))))
}
```

```r
eval(parse(text=sprintf('plot_aggregate <- grid.arrange(%s,nrow=2)',paste0(paste0(tablenames,'_plot_test'),collapse = ','))))
```

## NOTE : RED line is OBSERVATION, BLUE line is PREDICTION

```r
model_metric <- NULL
for(i in tablenames){
    eval(parse(text=sprintf('tahmin <- predict(%s_model,testing(%s_split))$.pred ; obs <- testing(%s_split)$visitor_num',i,i,i)))
    eval(parse(text=sprintf('fark <- abs(tahmin - obs); ape <- fark/obs ;%s_mape <- mean(ape)',i)))
    eval(parse(text=sprintf('newrow <- data.frame(channel = "%s", test_set_mape = %s_mape)',i,i)))
    
    model_metric <- rbind(model_metric,newrow)
}
```

```r
model_metric
```

### General MAPE

```r
model_metric_all <- NULL
for(i in tablenames){
    eval(parse(text=sprintf('tahmin <- predict(%s_model,testing(%s_split))$.pred ; obs <- testing(%s_split)$visitor_num',i,i,i)))
    eval(parse(text=sprintf('newrow <- data.frame(channel = "%s", obs = obs, pred = tahmin)',i)))
    
    model_metric_all <- rbind(model_metric_all,newrow)
}
```

```r
model_metric_all %>% 
mutate(ape = abs(pred-obs)/obs) %>% 
summarise(MAPE = mean(ape))
```
